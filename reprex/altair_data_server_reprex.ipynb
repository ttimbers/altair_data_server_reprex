# Altair data server reprex



```python
import pandas as pd
import altair as alt

# reduces the size of the notebook when run locally on a laptop or a JupyterHub when
# installing this version of altair-data-server: pip install -U git+https://github.com/yuvipanda/altair_data_server@better-hub-support-2
alt.data_transformers.enable('data_server')
```

```python
marathon_small = pd.read_csv("data/marathon_small.csv")
marathon_small
```

```python
# Run this cell to create a scatterplot of BMI against the time it took to run 10 km.

plot = (
    alt.Chart(marathon_small)
    .mark_point()
    .encode(
        x=alt.X("bmi", title="Body Mass Index"),
        y=alt.Y("km10_time_seconds", title="10 km run time (seconds)"),
    )
    .properties(
        width=400, height=400
    )  # to set the width and the height of the plot to a reasonable size
    .configure_axis(
        labelFontSize=20, titleFontSize=20
    )  # to set the font size for the axis labels and the axis titles
)
plot
```
